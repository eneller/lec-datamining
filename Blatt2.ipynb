```python
import scipy.cluster.hierarchy as hi
import matplotlib.pyplot as plt
import pickle as pk
data = pk.load(open("data.pickle", "rb"))
colors = [""]
```

```python
methods = ["single", "complete", "median", "weighted", "average"]
method_names = ["Single", "Complete", "Median", "Unweighted Avg.", "Group Avg."]

rows = cols = 5
gridspec={'wspace': 0.3, 'hspace': 0.3}
fig, axes= plt.subplots(rows,cols, figsize=(15,15))


for j,met in enumerate(methods):
    tree = hi.linkage(data, method= met, optimal_ordering=True)
    for i in range(1,6):
        labels = hi.fcluster(tree, i, criterion="maxclust")
        ax = axes[i-1,j]

        ax.scatter(data[:,0], data[:,1],c=labels,cmap="Accent" )
        # top label
        if i == 1:
            ax.set_title(method_names[j])

        # left label
        if j == 0:
            ax.set_ylabel('$k = %d$' % i)
```
