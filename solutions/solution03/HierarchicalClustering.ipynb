# Hierarchical Clustering

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram, centroid
from matplotlib import pyplot as plt, font_manager
import sys
sys.path.append('../../fonts')
from set_font import set_font
```

```python
data = np.array([
    [1, 1],
    [3, 2],
    [1.5, 5],
    [2, 6],
    [3, 4]
])
labels = [r'$\mathbf{p}_%d$' % i for i in range(1, len(data) + 1)]

set_font()
plt.rcParams['mathtext.bf'] = 'Libertinus Serif:italic:bold'  # Point notation should be bold and italic

def calc_linkage(method):
    tree = linkage(data, method=method, optimal_ordering=True)
    dendrogram(tree, labels=labels, color_threshold=0, link_color_func=lambda k: 'C0', leaf_font_size=plt.rcParams['font.size']);
    plt.xlabel('Data points')
    plt.ylabel('Distance')
    #plt.savefig('../assignment03/figures/HierarchicalClustering' + method.title() + '.pdf', bbox_inches='tight')
    
    print('Merge distances: ' + str(np.round(tree[:, 2], 2).astype(str)))
```

## Part 1: Single-Linkage

```python
calc_linkage('single')
```

## Part 2: Complete-Linkage

```python
calc_linkage('complete')
```

## Part 3: Centroid Linkage

```python
calc_linkage('centroid')
```
